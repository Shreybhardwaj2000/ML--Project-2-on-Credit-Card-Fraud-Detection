# Credit card fraud detection

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import accuracy_score
```

## Loading our data

```python
df1=pd.read_csv('creditcard.csv')
df1.head(10)
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
df1['Class'].value_counts()
```

## Seperating data for analysis

```python
legit=df1[df1.Class==0]
fraud=df1[df1.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df1.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=492)
```

```python
# now lets merge this sample dataset with fraud dataset
df2=pd.concat([legit_sample, fraud], axis=0)
df2.head()
```

```python
df2.tail()
```

```python
df2.groupby('Class').mean()
```

## Splitting the data into features(X) and targets(Y)

```python
X=df2.drop(columns='Class', axis=1)
Y=df2['Class']
```

```python
# Splliting our data into training and test
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(len(X_train))
print(len(X_test))
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
```

```python
# Accuracy on train data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_accuracy
```

```python
# Accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```
